```python
"""
about inplace = True

It is same as doing like this
df = df.set_index('Random')

if you give inplace = True, then the new dataframe which is returned by set_index() method i.e, 
    df.set_index('Random',inplace = True)
will be collected in your dataframe 'df'
"""
df.set_index('Random',inplace = True)
```

# iloc = integer location


dataframe.iloc[row_index,column_index]

```python
#np arrays
random = np.random.randint(low = 1, high =95, size = 20)
name = np.random.choice(['CSK', 'RCB', 'MI', 'KKR', 'RR'],size =  20)
choice = np.random.choice([7,8,9,10,11],size = 20)

#creating dictionary
dict1 = {
    'Random':random, 
    'Name':name,
    'Choice':choice
}

#creating dataframe
df = pd.DataFrame(dict1)
df
```

```python
# to access the first element in Random Column
df.iloc[0,0]
```

```python
# to access the 5th element in Choice Column
df.iloc[4,2]
```

```python
# to access the 11th element in the name column
df.iloc[11,1]
```

```python
# similarly you can do the slicing here as well

# to get first 5 rows 
df.iloc[0:5]
```

```python
# to get first 10 rows and first 2 columns
df.iloc[:10,:2]  #start:stop:step
```
