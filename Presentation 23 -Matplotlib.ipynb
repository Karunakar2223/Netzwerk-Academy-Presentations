# Scatter plot.

-->This plot is used for, Pairs of numerical data, with one variable on each axis, to look for a relationship between them.If the variables are correlated,the points will fall aalong a line or curve. 

-->plt.scatter()

```python
#creating two lists
x=np.arange(5,15)
y=np.arange(25,35)

#to create a scatter plot
plt.scatter(x,y,c = 'r',marker = '*')
#xlabel
plt.xlabel('X axis')
#ylabel
plt.ylabel('Y axis')
#title of the graph
plt.title('Scatter plot')
#to see the graph
plt.show()
#if you want you can save the graph
# plt.savefig('Test.png')
```

```python
print(x)
print(y)
```

```python
# sine wave from 0 to 2 pie
x1 = np.arange(0,2*np.pi,0.2)
y1 = np.sin(x1)

plt.plot(x1,y1,'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine wave')
plt.show()
```

```python
# cosine wave from 0 to 2 pie
x1 = np.arange(0,2*np.pi,0.2)
y1 = np.cos(x1)

plt.plot(x1,y1,'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Cosine wave')
plt.show()
```

# Bar graph.

--> A bar graph is a chart that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

-->plt.bar()

```python
# Roles
x = ['Analyst','Data Engineer','Machine learning Engineer'] 
# strength
y = [50,75,95]  

plt.bar(x, y)
plt.xlabel('Roles')
plt.ylabel('Total no. of Data Scientists')
plt.title('Bar Graph') 
plt.show()
```

# Histogram.

-->It is a bar graph like representation of data that buckets a range of outcomes into colums along the x-axis.The y-axis represents the number count in the data for each column and can be used to visualize data distributions.

-->plt.hist()

```python
x = np.array([22,87,5,43,56,8,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(x) 
plt.xlabel('X values')
plt.ylabel('No. of values')
plt.title("histogram") 
plt.show()
```

```python
x = np.array([22,87,5,43,56,8,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(x,bins = 10) 
plt.xlabel('X values')
plt.ylabel('No. of values')
plt.xticks(np.arange(0,90,5))
plt.title("histogram") 
plt.show()
```

# Pie chart.

-->A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

-->plt.pie

```python
# Data to plot
x = [1500, 800, 500, 90]
labels = ['Data scientists', 'Analysts', 'Machine Learning Engineer', 'Python Developers']
colors = ['r', 'g', 'b', 'y']

explode = (0, 0.2, 0, 0)  # explode 1st slice

# Plot
plt.pie(x, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%',shadow=True)
plt.title('Total number of Engineers')
plt.show()
```

# Creating Subplots
plt.subplot(nrows, ncols, index)

```python
# Ex : to create 4 plots

x = np.arange(10)
y = x**2

# to change the fig size
plt.figure(figsize = (10,10))

# first plot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.title('fig 1')

#second plot
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.title('fig 2')


#third plot
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.title('fig 3')


#fourth subplot
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.title('fig 4')


plt.show()
```
